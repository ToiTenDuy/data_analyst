```python
#Tạo bộ dữ liệu từ dữ liệu thu thập được
df = pd.DataFrame(list(zip(list00, list01, list02, list03,list04,list05,list06,list07,list08,list09, list010, list011, list013, list014, list015)),columns = ["brand", "model", "condition", "warranty", "screen_size", "processor", "ram", "card", "hard_driver", "made_in", "information", "price", "ssd", "hdd", "inch"])
```

```python
#Tách cách chuỗi trong processor thành 2 cột khác
df[['processor_brand', 'processor_name']] = df['processor'].str.split(' ', 1, expand=True)
```

# Xử lý sơ bộ dữ liệu, chỉnh sửa, thêm, xóa dữ liệu cần thiết

```python
#Tách cách kí tự thừa trong price
for value in df['price']:
    text = ' '.join(value)
    cleaned_text = text.replace('[', '').replace(']', '').replace(' đ', '')
    df[['price1']] = cleaned_text
```

```python
df['price'] = df['price1']
```

```python
df['made_in'] = df['made_in'].replace('Nước khác', np.nan)
```

```python
df['price'] = df['price'].str.replace('.', '', regex=True)
```

```python
df['price'] = df['price'].astype(float)
```

```python
df['condition'] = df['condition'].str.replace(r'\(chưa sửa chữa\)', '', regex=True)
```

```python
df['hard_driver'] = df['hard_driver'].str.extract('(\d+)')
```

```python
df = df.rename(columns={'hard_driver':'hard_driver(GB)'})
```

```python
df['ram'] = df['ram'].str.extract('(\d+)')
```

```python
df = df.rename(columns={'ram':'ram(GB)'})
```

```python
df['screen_size'] = df['inch']
```

```python
df = df.drop('information', axis=1)
```

```python
df = df.drop('price1', axis=1)
```

```python
# Sắp xếp lại các cột
df =df.reindex(columns=['brand', 'model', 'processor_brand', 'processor_name', 'os', 'screen_size', 'ram(GB)', 'card', 'ssd', 'hdd','hard_driver(GB)', 'made_in', 'warranty', 'condition', 'price'], level='brand')
```

```python
for brand in df['brand']:
    if brand == 'Apple':
        df['os'] = 'Macos'
    else:
        df['os'] = 'Windows'
```

```python
df['made_in'] = df['made_in'].replace('Đang cập nhật', np.nan)
```
